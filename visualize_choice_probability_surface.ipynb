```python
# Plotting
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns

# Data and numerics
import pandas as pd
import numpy as np
import scipy as scp

# System
import os 
from datetime import datetime

# Tensorflow
import tensorflow as tf
from tensorflow import keras

# Own
import make_data_wfpt as mdw
```

```python
# Define Model path:
model_path  = os.getcwd() + '/keras_models/dnnregressor_choice_probabilities_analytic_09_03_18_17_28_21/model_0'
ckpt_path = os.getcwd() + '/keras_models/dnnregressor_choice_probabilities_analytic_09_03_18_17_28_21/ckpt_0_' + 'final'

# Load model
model = keras.models.load_model(model_path)
model.load_weights(ckpt_path)
```

```python
# Generate data -- choice probabilities 
v_vals = np.arange(-3, 3, 0.1)
a_vals = np.arange(0.1, 3, 0.1)
data = []
for v in v_vals:
    for a in a_vals:        
        dataset,_, __ = mdw.make_data_choice_probabilities(v_range = [v,v], # uniform [0.6, 0.6]
                                      a_range = [a, a], # unifor
                                      w_range = [0.5, 0.5],  # uniform
                                      n_samples = 1,
                                      write_to_file = False)
        data.append(dataset)
        
dataset = pd.concat(data)
```

```python
features, labels, __, ___ = mdw.train_test_split_choice_probabilities(data = dataset,
                                                                      p_train = 1,
                                                                      write_to_file = False,
                                                                      from_file = False,
                                                                      backend = 'keras')
```

```python
# Predict choice probability with model
pred_choice_p = model.predict(features)
dataset['pred_choice_p'] = pred_choice_p
```

```python
# Predict
plt.rcParams['figure.figsize'] = [15.0, 10.0]
fig = plt.figure()
#fig, (ax1, ax2) = plt.subplot(121)
# ax1 = fig.gca(projection = '3d')
# ax2 = fig.gca(projection = '3d')
ax1 = fig.add_subplot(1,1,1, projection = '3d')
# ax2 = fig.add_subplot(1,2,2, projection = '3d')

# #ax.plot_trisurf(data['x'], data['y'], data['z'], cmap = plt.cm.viridis, linewidth = 0.2, alpha = 0.1)
ax1.plot_trisurf(dataset['v'], dataset['a'], dataset['pred_choice_p'], linewidth = 0.2, alpha = 0.2, color = 'orange')
ax1.set_ylabel('barrier_distance',  fontsize = 'x-large')
ax1.set_xlabel('drift',  fontsize = 'x-large')
ax1.set_zlabel('probability of hitting lower barrier',  fontsize = 'x-large')
ax1.set_title('Neural Network Predictions', fontsize = 'xx-large')
# ax2.plot_trisurf(dataset['v'], dataset['a'], dataset['p_lower_barrier'], linewidth = 0.2, alpha = 0.5, color = 'blue')
# ax2.set_ylabel('barrier_distance',  fontsize = 'x-large')
# ax2.set_xlabel('drift',  fontsize = 'x-large')
# ax2.set_zlabel('probability of hitting lower barrier',  fontsize = 'x-large')
# ax2.set_title('Analytic Solution', fontsize = 'xx-large')
```
