```python
# Try my own data loader

bs = 40000

# Load data
data_folder = os.getcwd() + '/data_storage/kde/weibull/train_test_data'

# rt_choice
train_f, train_l, test_f, test_l = kde_load_data(folder = data_folder)
train_f = train_f.values
test_f = test_f.values
```

```python
train_f = train_f.astype(np.float32, copy = False)
train_l = train_l.astype(np.float32, copy = False)
test_f = test_f.astype(np.float32, copy = False)
test_l = test_l.astype(np.float32, copy = False)
```

```python
train_f, train_l, test_f, test_l = map(
    torch.from_numpy, ((train_f, train_l, test_f, test_l)))
```

```python
train_f = torch.as_tensor(train_f, dtype = torch.float32)
test_f = torch.as_tensor(test_f, dtype = torch.float32)
train_l = torch.as_tensor(train_l, dtype = torch.float32)
test_l = torch.as_tensor(test_l, dtype = torch.float32)
```

```python
train_ds = TensorDataset(train_f, train_l)
test_ds = TensorDataset(test_f, test_l)
```

```python
train_dl = DataLoader(train_ds, batch_size = bs)
test_dl = DataLoader(test_ds, batch_size = bs * 2)
```

```python
def get_model_2():
    model = nn.Sequential(OrderedDict([
                      ('lin1', nn.Linear(8, 10)),
                      ('relu1', nn.ReLU()),
                      ('lin2', nn.Linear(10, 1))]))
    return model, optim.SGD(model.parameters(), lr = lr)
```

```python
from IPython.core.debugger import set_trace

lr = 0.5
epochs = 10

loss_func = F.mse_loss

model, opt = get_model_2()

for epoch in range(epochs):
    model.train()
    cnt = 0
    for xb, yb in train_dl:
        loss = loss_func(model(xb), yb)
        loss.backward()
        opt.step()
        opt.zero_grad()
        cnt += 1
        print(cnt * bs)
        
    model.eval()
    with torch.no_grad():
        valid_loss = sum([loss_func(model(xb), yb) for xb, yb in test_dl])
            
    print(epoch, valid_loss / len(test_dl))           
```

```python
state = {'epoch': epoch,
         'state_dict': model.state_dict(),
         'optimizer': opt.state_dict()
        }
```

```python
# Save and Load model
PATH = os.getcwd() + '/torch/model_checkpoint.pt'
torch.save(state, PATH)
```

```python
state = torch.load(PATH)
model.load_state_dict(state['state_dict'])
opt.load_state_dict(state['optimizer'])
```
