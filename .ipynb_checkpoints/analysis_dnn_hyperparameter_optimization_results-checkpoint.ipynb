```python
# Load packages
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import os
import scipy as scp
import scipy.stats as scps
from glob import glob
# Load my own functions
import dnnregressor_train_eval_keras as dnnk
import make_data_wfpt as mdw
```

```python
# Load hyperparameter optimization results 
cwd = os.getcwd()
directory = cwd + '/keras_models/dnnregressor_wftp_hyp_opt_choice_rt_09_16_18_21_23_13/'
model_specification_file = 'model_specifications.csv'
model_specifications = pd.read_csv(directory + model_specification_file)
history_filenames = glob(directory + 'history_*')
```

```python
def get_model_num_from_history_csv(filename = 'akldlaksjds20...'):
    my_digits = []
    chars = filename[-7:-3]
    for char in chars:
        if char.isdigit():
            my_digits.append(char)
    return int(''.join(my_digits))
```

```python
get_model_num_from_history_csv(filename = history_filenames[0])
```

```python
histories = []
for filename in history_filenames:
    histories.append(pd.read_csv(filename))
    histories[-1]['model_cnt'] = get_model_num_from_history_csv(filename = filename)
```

```python
model_specifications['val_mean_squared_error'] = 0
model_specifications['val_mean_squared_error_ckpt'] = 'final'

for history in histories:
    model_cnt_tmp = history['model_cnt'][0]
    model_specifications.loc[model_specifications['model_cnt'] == model_cnt_tmp, 'val_mean_squared_error'] = \
    list(history['val_mean_squared_error'])[-1]
#     model_specifications.loc[model_specifications['model_cnt'] == model_cnt_tmp, 'val_mean_squared_error_ckpt'] = \
#     history['val_mean_squared_error'].idxmin() + 1
```
