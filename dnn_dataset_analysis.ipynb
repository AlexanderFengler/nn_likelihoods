```python
# Load packages 
import numpy as np
import pandas as pd
import scipy as scp
import matplotlib.pyplot as pyplot
import os
import re
```

```python
# Load dataset
data_directory = os.getcwd() + '/data_storage'
file_list = os.listdir(data_directory)
```

```python
# Specify datatype and number of data points to identify file to open
datatype = '_from_simulation_mix_'
n = 1000000

# Search for matching file in directory
my_regex = '^data_' + str(n) + datatype + '*'
p = re.compile(my_regex)

file_to_open = []
for file in file_list:
    if p.search(file):
        file_to_open.append(file)
        
# Load file
data = pd.read_csv(data_directory + '/' + file_to_open[0], delimiter= ',')
# Drop unwanted index column
data = data.drop(columns = data.keys()[0])
```

```python
data2 = data.copy()
```

```python
data_full = pd.concat([data2, data])
```
