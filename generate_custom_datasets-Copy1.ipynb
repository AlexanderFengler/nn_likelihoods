```python
import tensorflow as tf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy as scp
import os
from datetime import datetime

# My modules
import make_data_wfpt as mdw
```

```python
# Make dataset that concatenates 
v_vals = [0.0]
a_vals = [1, 1.5, 2, 2.5, 3]
w_vals = [0.25, 0.5, 0.75, 1]
n_samples = 200000
cwd = os.getcwd()
file_directory = cwd + '/data_storage/' + 'data_' + str(n_samples) + '_from_simulation_mix_param_grid_' + \
                 datetime.now().strftime('%m_%d_%y_%H_%M_%S') + '.csv'


data = []
for v_val in v_vals:
    for w_val in w_vals:
        for a_val in a_vals:
            data_tmp, __, ___ = mdw.make_data_rt_choice(v_range = [v_val, v_val],
                                                        a_range = [a_val, a_val],
                                                        w_range = [w_val, w_val],
                                                        method = 'sim',
                                                        n_samples = n_samples,
                                                        write_to_file = False,
                                                        mixture_p = [0.9, 0.05, 0.05],
                                                        print_detailed_cnt = False
                                                       )
            data.append(data_tmp)

data = pd.concat(data)
data = data.sample(frac = 1).reset_index(drop = True)
data.to_csv(file_directory)

# Split into train and test set
_, __, ___, ____ = mdw.train_test_split_rt_choice(p_train = 0.8, 
                                                  write_to_file = True,
                                                  f_signature = '_from_simulation_mix_param_grid_',
                                                  n_samples = n_samples)
```
